# Line Chart

```python
import matplotlib.pyplot as plt
import numpy as np
x=[1,3,5,7,9]
y=[1,3,5,7,9]
plt.plot(x,y) # Plotting graph
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(0.0,6.0,0.01)
y=[x**2 for x in x]
plt.plot(x,y)
plt.show()
```

```python
# multiple lines on the same graph
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,10)
y1=x+2
y2=x*6
y3=x/5
plt.plot(x,y1,x,y2,x,y3)
plt.show()
```

```python
#Adding grids,axes,labels,markers on a multiple line graph
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,10)
y1=x+2
y2=x*6
y3=x/5
plt.plot(x,y1,'g--',label='y1')
plt.plot(x,y2,color="pink",label='y2')
plt.plot(x,y3,'r-o',label='y3')
plt.grid(True)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
#Adding grids,axes,labels,markers on a multiple line graph
import matplotlib.pyplot as plt
import numpy as np
x=np.arange(1,10)
y1=x+2
y2=x*6
y3=x/5
plt.plot(x,y1,'r--',label='y1')
plt.plot(x,y2,color="green",label='y2')
plt.plot(x,y3,'b-o',label='y3')
plt.grid(True)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend(bbox_to_anchor=(1, 1)) # Moving legend to outside of graph
plt.show()
```

```python
import matplotlib.pyplot as plt
year=[1960,1970,1980,1990,2000,2010]
pop_us=[44.9,45.23,55.23,56.21,57.2,59.3]
pop_ind=[441.9,452.23,553.23,564.21,575.2,596.3]
plt.figure(figsize=(10,10)) # figure size
plt.plot(year,pop_us,color='green')
plt.plot(year,pop_ind,color='orange')
plt.xlabel('countries')
plt.ylabel('year')
plt.title('India population vs US population over different years',color='b')
plt.show()
```

# Bar Chart

```python
import matplotlib.pyplot as plt
height=[3,12,45,78]
bars=('a','b','c','d')
plt.bar(bars,height)
plt.title('Bar Graph')
plt.xlabel('bars')
plt.ylabel('height')
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.bar([1,2,3],[5,2,1])
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title('simple bar chart')
plt.show()
```

```python
import matplotlib.pyplot as plt
student_name=["Jhon","Sahil","Monica","Lara"]
marks=[75,98,53,70]
plt.bar(student_name,marks)
plt.xlabel('Student_name')
plt.ylabel('Marks')
plt.title('simple bar chart')
plt.show()
```

# Pie Chart

```python
import matplotlib.pyplot as plt
languages=["Python","Java","C++"]
percentage=[60,30,10]
plt.pie(percentage,labels=languages)
plt.title("simple Pie-Chart")
plt.show()
```

```python
# More Options in Pie chart
import matplotlib.pyplot as plt
labels='python','c++','java','ruby'
sizes=[90,50,40,110]
colors=['gold','grey','green','blue']
explode=(0,0,0.2,0)
plt.pie(sizes,explode=explode,labels=labels,colors=colors,autopct='%1.1f%%',shadow=True)
plt.axis('equal')
plt.show()
```

# Subplots

```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,5,11)
y=x**2
fig,(ax1,ax2)=plt.subplots(ncols=2,nrows=1)
ax1.plot(x,y,'m')
ax2.plot(y,x,'g')
ax1.set_xlabel('x-axis')
ax1.set_ylabel('y-axis')
ax2.set_xlabel('x-axis')
ax2.set_ylabel('y-axis')
ax1.set_title('parabola')
ax2.set_title('inversed parabola')
plt.show()
```

```python
# saving a figure
fig.savefig('sub.jpeg')
```

```python
import matplotlib.pyplot as plt
import numpy as np
fig,(ax1,ax2,ax3)=plt.subplots(3,1)
x=np.linspace(0,5,10)
y=x**2
ax1.plot(x,y)
ax2.plot(x,y)
ax3.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)


t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

plt.figure()
plt.subplot(211)
plt.plot(t1, f(t1), color='tab:blue', marker='o')
plt.plot(t2, f(t2), color='black')

plt.subplot(212)
plt.plot(t2, np.cos(2*np.pi*t2), color='tab:orange', linestyle='--')
plt.show()
```
