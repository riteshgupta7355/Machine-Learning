<h1><center>Seaborn</center></h1>

<b>Seaborn package is built on matplotlib but provides high level interface for drawing attractive statistical graphics, It specifically targets statistical data visualization.</b>

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings("ignore")
%matplotlib inline
tip=sns.load_dataset('tips')
sns.set()
tip.head()
```

```python
tip.describe()
```

```python
plt.hist(tip.total_bill)
plt.plot()
plt.show()
```

# Distribution Plot

<b>Distribution plot will gives univariate. In histogram on top of that line is known as Kernel Density Function(KDE)<b>

```python
sns.distplot(tip.total_bill,bins=10)
plt.show()
```

```python
sns.distplot(tip.total_bill,bins=10,kde=False)
plt.show()
```

# Joint Plot

```python
sns.jointplot(x='total_bill',y='tip',data=tip,kind='scatter',size=5)
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tip,kind='hex',size=5)
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tip,kind='reg',size=5)
plt.show()
```

```python
sns.jointplot(x='total_bill',y='tip',data=tip,kind='kde',size=5)
plt.show()
```

# Pair Plot

<b>Pair plot will plot a diagram all against the all possible combination of all numerical values. This  is most important plots

We will work on this plot through  out our ML.</b>


```python
sns.pairplot(data=tip,size=2)
plt.show()
```

```python
sns.pairplot(data=tip,kind='scatter',hue='sex',size=3)
plt.show()
```

```python
sns.pairplot(data=tip,kind='reg',hue='sex',size=5)
plt.show()
```

# Rug Plot

```python
sns.rugplot(tip.total_bill)
plt.show()
```

# Bar Plot

```python
sns.barplot(x='sex',y='total_bill',data=tip)
plt.show()
```

# Count Plot

```python
sns.countplot(x='sex',data=tip)
plt.show()
```

# Box Plot

<b>Box plot  is used to shown the distribution of categorical data and quantitative data in a way that comparison between two variables.


Box plot shows the quartiles of the dataset while the whiskers extend to show the rest of the distribution, except for points that are determined to be ‘outliers’.</b>


```python
sns.boxplot(data=tip)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(data=tip,x='day',y='total_bill',hue='sex',palette='rainbow')
plt.legend(bbox_to_anchor=(1.1,1.1))
plt.show()
```

```python
sns.boxplot(data=tip,palette='coolwarm',orient='h')
plt.show()
```

# violinplot

```python
plt.figure(figsize=(10,5))
sns.violinplot(data=tip,x='day',y='total_bill',hue='smoker',palette='rainbow')
plt.legend(bbox_to_anchor=(1.12,1.02))
plt.show()
```

# Strip Plot

```python
sns.stripplot(x='day',y='total_bill',data=tip)
plt.show() 
```

```python
sns.stripplot(x='day',y='total_bill',data=tip,jitter=True)
```

```python
sns.stripplot(x="sex", y="total_bill", hue="day",data=tip, jitter=True)
plt.show()
```

```python
sns.swarmplot(x="day", y="total_bill", data=tip)
plt.show()
```

```python
sns.factorplot(x='sex',y='total_bill',data=tip)
plt.show()
```

# Grids

# Pair Grid

<b> Pair Grid is a subplot grid for plotting pairwise relationships in a data set</b>

```python
g=sns.PairGrid(tip)
g=g.map(plt.scatter)
```

```python
g=sns.PairGrid(tip)
g=g.map(plt.scatter)
g=g.map_diag(plt.hist)
```

```python
g=sns.PairGrid(tip)
g=g.map(sns.kdeplot)
```

```python
g=sns.PairGrid(tip)
g=g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
plt.show()
```

# Facet Grid

<b>Facet Grid is the general way to create grids of plot based on a feature</b>

```python
g=sns.FacetGrid(tip,col='time',row='smoker')
g=g.map(plt.hist,'total_bill')
```

```python
g=sns.FacetGrid(tip,col='time',row='smoker',hue='sex')
g=g.map(plt.hist,'total_bill').add_legend()
```

```python
g=sns.FacetGrid(tip,col='time',row='smoker',hue='sex')
g=g.map(plt.scatter,'total_bill','tip').add_legend()
```
